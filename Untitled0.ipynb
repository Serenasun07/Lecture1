```python
import pandas as pd
from google.colab import drive
import matplotlib.pyplot as plt
```

```python
drive.mount('/content/gdrive/', force_remount=True)
```

```python
df = pd.read_csv("/content/gdrive/MyDrive/ECON441B/^VIX.csv")
```

```python
df.set_index("Date", inplace = True)
```

```python
df.index = pd.to_datetime(df.index)
```

```python
plt.figure(figsize = (12,5))
plt.plot(df["Close"])
plt.title("VIX INDEX")
plt.xlabel("Date")
plt.ylabel("VIX")
plt.grid()
plt.show()
```

In the past year, there was a stock price lift from february to march, and reached the highest value in early March. From October to date, there is a descending trend in prices.
